```python
import pandas
import numpy as np
from sklearn import linear_model, datasets
from sklearn.preprocessing import normalize
import tensorflow as tf
```

```python
data_source = "http://goo.gl/vhm1eU"
names = ['prg', 'plas', 'pres','skin','test','mass','pedi', 'age','class']

df = pandas.read_csv(data_source, names=names)
df.ix[:5]
len(df)
```

```python
data=df[:768].as_matrix()
data[:, :8] = normalize(data[:, :8], axis=0)
data[:5]
```

```python
x_data = data[:, :8]
x_data = np.insert(x_data,0,1,axis=1).transpose()
y_data = data[:, -1]
x_data.shape
```

```python
X = tf.placeholder(tf.float32) #input 느낌
Y = tf.placeholder(tf.float32)

Wm=tf.Variable(tf.random_uniform([1,len(x_data)],-1.0,1.0))
```

```python
h =tf.matmul(W,X)
hypothesis =tf.div(1.,1.+tf.exp(-h))
cost =-tf.reduce_mean(Y*tf.log(hypothesis )+(1-Y)*tf.log(1-hypothesis))

a = tf.Variable(0.01)
optimizer = tf.train.GradientDescentOptimizer(a)
train = optimizer.minimize(cost)
```

```python
init =tf.initialize_all_variables()

sess=tf.Session()
sess.run(init)
```

```python
for step in range(20000000):
    sess.run(train, feed_dict={X:x_data, Y:y_data})
    if step % 50000 == 0:
        print (step, sess.run(cost, feed_dict={X:x_data, Y:y_data}),sess.run(W))
```

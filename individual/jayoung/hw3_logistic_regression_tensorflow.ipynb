```python
import pandas
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
%matplotlib inline
```

```python
#https://archive.ics.uci.edu/ml/machine-learning-databases/pima-indians-diabetes/pima-indians-diabetes.names

data_source = "http://goo.gl/vhm1eU"
names = ['prg', 'plas', 'pres','skin','test','mass','pedi', 'age','class']

df = pandas.read_csv(data_source, names=names)
#df.shape
```

```python
y_data = df["class"].values
```

```python
df_norm = (df.ix[:,:-1] - df.ix[:,:-1].min()) / (df.ix[:,:-1].max() - df.ix[:,:-1].min())
x_data = df_norm[:].values

x_o=np.ones(len(x_data)).reshape(-1,1)

f_x_data=np.float32(np.concatenate((x_o, x_data),axis=1))
t_f_x_data=f_x_data.transpose()
t_f_x_data.shape
```

```python
W = tf.Variable(tf.random_uniform([1,len(f_x_data[0])], -1.0, -1.0))

h = tf.matmul(W,t_f_x_data)
hypothesis = tf.div(1., 1. + tf.exp(-h))
cost = -1 * tf.reduce_mean(y_data * tf.log(hypothesis) + (1-y_data) * tf.log(1-hypothesis))

a = tf.Variable(0.01)
optimizer = tf.train.GradientDescentOptimizer(a)
train = optimizer.minimize(cost)

init = tf.initialize_all_variables()

sess = tf.Session()
sess.run(init)

for step in range(10000000):
    sess.run(train)
    if step % 100000 == 0:
        print (step, sess.run(cost), sess.run(W))
        
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(z))

def hypothesis_function(x, theta):
    z = (np.dot(-x,theta))    
    return sigmoid(z)

final_W=sess.run(W)
y_data=y_data.reshape(len(y_data),1)
#print(len(y_data))

sum((hypothesis_function(f_x_data, final_W.T)> 0.5) == y_data) / y_data.shape[0]
```
